# Import Library's

```python
import torch
import os
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import kagglehub

from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path = kagglehub.dataset_download("svaningelgem/crypto-currencies-daily-prices")
print("Path to dataset files:", path)
```

### CryptoData
A class designed to hold the Dataset, was originally designed to hold images, is unnessesary, but I didn't want to get rid of it. 


```python
class CryptoData(Dataset):
    def __init__(self, datadir):
        self.data = pd.concat(
            [pd.read_csv(os.path.join(datadir, f)) for f in os.listdir(datadir) if f.endswith('.csv')], ignore_index=True)
    
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]
```

```python
dataset = CryptoData(datadir= path)
print(dataset.__len__())
```
